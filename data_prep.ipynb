```python
train = pd.read_csv("small_train.csv")
contractions = {
    "ain't": "isnt",
    "aren't": "are not",
    "can't": "cannot",
    "could've": "could have",
    "couldn't": "could not",
    "didn't": "did not",
    "doesn't": "does not",
    "don't": "do not",
    "hadn't": "had not",
    "hasn't": "has not",
    "haven't": "have not",
    "he'd": "he had",
    "he'll": "he will",
    "he's": "he is",
    "how'd": "how did",
    "how'll": "how will",
    "how's": "how is",
    "i'd": "i had",
    "i'll": "i will",
    "i'm": "i am",
    "i've": "i have",
    "isn't": "is not",
    "it'd": "it had",
    "it'll": "it will",
    "it's": "it is",
    "let's": "let us",
    "ma'am": "madam",
    "might've": "might have",
    "mightn't": "might not",
    "must've": "must have",
    "mustn't": "must not",
    "needn't": "need not",
    "shan't": "shall not",
    "she'd": "she had",
    "she'll": "she will",
    "she's": "she is",
    "should've": "should have",
    "shouldn't": "should not",
    "that's": "that is",
    "there's": "there is",
    "they'd": "they had",
    "they'll": "they will",
    "they're": "they are",
    "they've": "they have",
    "wasn't": "was not",
    "we'd": "we had",
    "we'll": "we will",
    "we're": "we are",
    "we've": "we have",
    "weren't": "were not",
    "what'll": "what will",
    "what're": "what are",
    "what's": "what is",
    "what've": "what have",
    "where's": "where is",
    "who'll": "who will",
    "who's": "who is",
    "won't": "will not",
    "would've": "would have",
    "wouldn't": "would not",
    "you'd": "you had",
    "you'll": "you will",
    "you're": "you are",
    "you've": "you have",
    "y'all": "you all"
}

def replace_contractions(text):
    for contr, repl in contractions.items():
        text = text.replace(contr, repl)
    return text
```

```python
def tokenize_question(row, lemmatizer):
    # tokenize using nlkt's method
    q1 = nltk.word_tokenize(row["question1"])
    q2 = nltk.word_tokenize(row["question2"])
    print(q1)
    q1 = lemmatizer(q1)
    print(q1)
    row["question1"] = " ".join(q1)
    print("----------------")
    q2 = lemmatizer(q2)
    row["question2"] = " ".join(q2)
    
    return row
```

```python


# make questions lowercase and remove punctuation
train["question1"] = train["question1"].str.lower()
train["question2"] = train["question2"].str.lower()

train["question1"] = train["question1"].apply(lambda text: replace_contractions(text))
train["question2"] = train["question2"].apply(lambda text: replace_contractions(text))

punct_table = str.maketrans('', '', string.punctuation)
train['question1'] = train['question1'].str.translate(punct_table).str.replace("\d+", "")
train['question2'] = train['question2'].str.translate(punct_table).str.replace("\d+", "")


# drop na 
train.dropna(inplace=True)
train = train[train["question1"]!=""]
train = train[train["question2"]!=""]


# lemmatize and tokenize
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatizer = np.vectorize(lemmatizer.lemmatize)

#punct_table = str.maketrans('', '', string.punctuation)
#train['question1'] = train['question1'].str.translate(punct_table)
#train['question2'] = train['question2'].str.translate(punct_table)

train = train.apply(lambda row: tokenize_question(row, lemmatizer), axis=1)
```
